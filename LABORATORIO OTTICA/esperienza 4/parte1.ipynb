```python
import Tiamat as tiam
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradtorad(gradi,primi):
    angolo = gradi + primi/60
    return (angolo*np.pi)/180

# Angoli già sfasati
theta = np.array([gradtorad(225,20),gradtorad(201,15),gradtorad(180,0),gradtorad(159,35),gradtorad(134,25)])
sintheta = np.sin(theta)
errsintheta = np.array([np.abs(np.cos(theta[0])*gradtorad(0,1)),
                        np.abs(np.cos(theta[1])*gradtorad(0,1)),
                        np.abs(np.cos(theta[2])*gradtorad(0,1)),
                        np.abs(np.cos(theta[3])*gradtorad(0,1)),
                        np.abs(np.cos(theta[4])*gradtorad(0,1))])

m = np.array([-2,-1,0,1,2])

slope,intercept,cov_matrix = tiam.fit_lineare(m,sintheta,err_R_wi=errsintheta)

lampda = 5893*10**(-10)

p = lampda/slope
print(p)
print("G:",(1/p)/1000)

plt.scatter(m,sintheta)
print(errsintheta)
plt.errorbar(m,sintheta,xerr=[0,0,0,0,0],yerr=errsintheta)
plt.grid()

print("sintheta: ",sintheta)
```

```python

sfas=gradtorad(94,30)
#celeste
thetai=np.array((gradtorad(85,30),gradtorad(68,40),gradtorad(50,40),gradtorad(101,55),gradtorad(119,40)))+sfas
mi=np.array((0,1,2,-1,-2))
print("indaco:",thetai)
plt.scatter(mi,np.sin(thetai),c="skyblue")

#blue
thetas=np.array((gradtorad(85,30),gradtorad(69,15),gradtorad(50,10),gradtorad(102,55),gradtorad(121,10)))+sfas
mc=np.array((0,1,2,-1,-2))
print("celeste:",thetas)
plt.scatter(mc,np.sin(thetas),color="b")
#verde
thetav=np.array((gradtorad(85,30),gradtorad(67,10),gradtorad(47,10),gradtorad(103,5),gradtorad(123,10)))+sfas
mv=np.array((0,1,2,-1,-2))
plt.scatter(mv,np.sin(thetav),color="g")
#rosso
mr=np.array((0,1,2,-1,-2))
thetar=np.array((gradtorad(85,30),gradtorad(62,40),gradtorad(34,50),gradtorad(108,40),gradtorad(137,20)))+sfas
plt.scatter(mr,np.sin(thetar),c="r")

#indaco
##4B0082
mindaco=np.array((0,1,2,-1,-2))
thetaindaco=np.array((gradtorad(85,30),gradtorad(69,40),gradtorad(53,15),gradtorad(101,15),gradtorad(117,35))) + sfas
plt.scatter(mindaco,np.sin(thetaindaco))

slope0,intercept0,cov_matrix0 = tiam.fit_lineare(mi,np.sin(thetai))
slope1,intercept1,cov_matrix0 = tiam.fit_lineare(mc,np.sin(thetas))
slope2,intercept2,cov_matrix0 = tiam.fit_lineare(mv,np.sin(thetav))
slope3,intercept3,cov_matrix0 = tiam.fit_lineare(mr,np.sin(thetar))
slope4,intercept4,cov_matrix0 = tiam.fit_lineare(mindaco,np.sin(thetaindaco))

plt.plot(mi,slope0*mi+intercept0,color="b")
plt.plot(mi,slope1*mi+intercept1,color="skyblue")
plt.plot(mi,slope2*mi+intercept2,color="g")
plt.plot(mi,slope3*mi+intercept3,color="r")
plt.plot(mi,slope4*mi+intercept4,color="#4B0082")

lampda0 = slope0*p
lampda1 = slope1*p
lampda2 = slope2*p
lampda3 = slope3*p
lampda4 = slope4*p

print("blue ",lampda0)
print("celeste: ",lampda1)
print("verde: ",lampda2)
print("rosso: ",lampda3)
print("indaco: ",lampda4)

plt.grid()
```
